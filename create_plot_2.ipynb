```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("merged_data/changed_merged_data.csv")

winner_data = data[data["Race Position"] == 1]
full_name = winner_data["Driver First Name"]+ " " + winner_data["Driver Last Name"]
wins_pivot = winner_data.pivot_table( index = full_name, columns="Year", values="Race Position", aggfunc="count", fill_value=0 )
plt.figure(figsize=(12, 6))
wins_pivot.plot(kind="bar", stacked=True, colormap="tab20", edgecolor="black", ax=plt.gca())
plt.title("Race Wins Per Year by Driver (2018-2023)", fontsize=14)
plt.xlabel("Driver", fontsize=12)
plt.ylabel("Number of Wins", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```
